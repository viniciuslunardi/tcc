# Iniciando limpeza dos dados e geração do dataset final 

```python
# removendo colunas com valores nulos para iniciar experimentos
df_cleaned = dataframe_ids.dropna(subset=["id_integer", "Quais as 5 principais métricas que você utiliza em seu(s) time(s) e/ou organização?"])
```

```python
# dataframe com as métricas normalizadas
dataframe_metrics = pd.read_excel('dados_brutos_ids_int.xlsx', sheet_name='MÉTRICAS DE-PARA')
```

```python
df_merged = pd.merge(df_cleaned, dataframe_metrics, how="left", left_on="id_integer", right_on="id_integer")
df_grouped = df_merged.groupby("id_integer")["PARA"].apply(lambda x: ', '.join(x.dropna())).reset_index()
df_final = pd.merge(df_cleaned, df_grouped, on="id_integer", how="left")
df_final.rename(columns={"PARA": "sanitized_metrics"}, inplace=True)
```

```python
column_renaming = {
    "Qual seu papel na equipe que atua?": "role",
    "Há quanto tempo (em anos) você trabalha com desenvolvimento de software?": "years_exp",
    "Aproximadamente quantos colaboradores tem sua organização?": "org_size",
    "Para quais áreas de domínio a sua organização desenvolve software?": "org_field",
    "Qual(is) métodos ágeis seu time utiliza?": "agile_methods",
    "Em quais cerimônias típicas de métodos ágeis você aplica gerência de riscos e/ou métricas? [Reunião de planejamento da sprint]": "use_metrics_planning",
    "Em quais cerimônias típicas de métodos ágeis você aplica gerência de riscos e/ou métricas? [Reunião de revisão da sprint]": "use_metrics_review",
    "Em quais cerimônias típicas de métodos ágeis você aplica gerência de riscos e/ou métricas? [Reunião semanal]": "use_metrics_weekly",
    "Em quais cerimônias típicas de métodos ágeis você aplica gerência de riscos e/ou métricas? [Reunião diária]": "use_metrics_daily",
    "Em quais cerimônias típicas de métodos ágeis você aplica gerência de riscos e/ou métricas? [Reunião de retrospectiva da sprint]": "use_metrics_retro",
    "Em quais cerimônias típicas de métodos ágeis você aplica gerência de riscos e/ou métricas? [Nenhuma das respostas acima]": "use_metrics_na",
    "Em quais dos itens abaixo você aplica métricas para o gerenciamento?": "metrics_category"
}

# removing columns from the dataframe
df_final.drop(columns=['Quais as 5 principais métricas que você utiliza em seu(s) time(s) e/ou organização?'], inplace=True, errors='ignore')
df_final = df_final[df_final['sanitized_metrics'].notna() & df_final['sanitized_metrics'].str.strip().astype(bool)]


# Rename the columns in the final DataFrame
df_final.rename(columns=column_renaming, inplace=True)
```

```python
# dataset oficial que será utilizado para experimentos e para o modelo final
df_final.to_excel("cleaned_dataset.xlsx", index=False)
```

```python
df_cleaned = pd.read_excel("cleaned_dataset.xlsx")

# Remove the incorrect metric "Métricas ágeis" from the "sanitized metrics" column
df_cleaned['sanitized_metrics'] = df_cleaned['sanitized_metrics'].str.replace(r'\bMétricas ágeis\b,?\s*', '', regex=True).str.strip(", ")

# Split the "sanitized metrics" column into multiple columns
split_metrics = df_cleaned['sanitized_metrics'].str.split(', ', expand=True)

# Rename the new columns as metric_1, metric_2, etc.
num_metrics = split_metrics.shape[1]
split_metrics.columns = [f'metric_{i+1}' for i in range(num_metrics)]

# Merge the expanded metrics columns back to the original DataFrame
df_final = pd.concat([df_cleaned, split_metrics], axis=1)

df_final.to_excel("cleaned_dataset_final.xlsx", index=False)
```
