```python
# Carregar a estrutura de colunas salva durante o treinamento
column_structure = joblib.load('column_structure.joblib')
```

```python
# Adicionar as colunas faltantes no user_df e preencher com 0
missing_cols = set(column_structure) - set(user_df.columns)
for col in missing_cols:
    user_df[col] = 0
```

```python

# Reordenar as colunas para garantir que a ordem seja a mesma do treino
user_df = user_df[column_structure]
```

```python
# Remover as colunas relacionadas às categorias de métricas (se presentes)
categories = ['Cronograma e progresso', 'Produto', 'Processo', 'Tecnologia', 'Pessoas', 'Cliente']
user_df = user_df.drop(columns=categories, errors='ignore')  # Remover categorias de métricas
```

```python
# Verificar se os tipos de dados estão corretos (inteiros/floats)
user_df = user_df.astype(float)
```

```python

# Prever as afinidades para cada usuário simulado
category_probabilities = {user_idx: {} for user_idx in range(len(user_df))}
```

```python
user_classification = {user_idx: [] for user_idx in range(len(user_df))}  # Para armazenar as categorias com > 50% de afinidade
```

```python
# Definir o threshold mínimo de afinidade para recomendação
threshold = 0.5
```

```python

for category in categories:
    # Carregar o modelo treinado
    model_filename = f"modelo_{category}_ajustado.joblib"
    rf_binary = joblib.load(model_filename)
    
    # Prever as afinidades para todos os usuários
    user_affinities = rf_binary.predict_proba(user_df)[:, 1]
    
    # Armazenar as afinidades e classificar as métricas com afinidade > 50%
    for user_idx, affinity in enumerate(user_affinities):
        category_probabilities[user_idx][category] = affinity
        if affinity >= 0.5:
            user_classification[user_idx].append(category)  # Classificar se afinidade > 50%

# Exibir as probabilidades de afinidade para cada usuário e suas classificações
for user_idx, probabilities in category_probabilities.items():
    print(f"\nUsuário {user_idx + 1}:")
    for category, probability in probabilities.items():
        print(f"Afinidade para {category}: {probability:.2f}")
    
    # Exibir as categorias de métricas com afinidade > 50%
    if user_classification[user_idx]:
        print(f"Usuário {user_idx + 1} está classificado com as seguintes métricas (> 50% afinidade): {', '.join(user_classification[user_idx])}")
    else:
        print(f"Usuário {user_idx + 1} não tem afinidade acima de 50% para nenhuma métrica.")
```

```python
metrics_data = pd.read_excel('extracted_metrics.xlsx', sheet_name="Extracted metrics")
metrics_data.head()
```

```python
category_to_metrics = metrics_data.groupby('Translation')['Metric Name'].apply(list).to_dict()
```

```python
# Função para recomendar métricas com base nas afinidades das categorias
def recomendar_metricas_com_afinidade(category_probabilities, category_to_metrics, threshold):
    recommendations = {}
    for user_idx, category_affinities in category_probabilities.items():
        user_recommendations = []
        for categoria, afinidade in category_affinities.items():
            if categoria in category_to_metrics and afinidade >= threshold:
                # A afinidade de cada métrica é igual à afinidade da sua categoria
                for metrica in category_to_metrics[categoria]:
                    if pd.notna(metrica):  # Certificar-se de que a métrica não é 'nan'
                        user_recommendations.append((metrica, afinidade))
        
        # Ordenar métricas por afinidade (decrescente) para cada usuário
        user_recommendations.sort(key=lambda x: x[1], reverse=True)
        recommendations[user_idx] = user_recommendations
    
    return recommendations
```

```python
# Exemplo de uso da função de recomendação:
metric_recommendations_with_affinity = recomendar_metricas_com_afinidade(category_probabilities, category_to_metrics, threshold)
```

```python
# Exibir as recomendações de métricas para cada usuário, mostrando também as afinidades e evitando duplicatas
for user_idx, metrics_with_affinity in metric_recommendations_with_affinity.items():
    print(f"\nRecomendações de Métricas para Usuário {user_idx + 1} (com afinidade >= {threshold}):")
    if metrics_with_affinity:
        seen_metrics = set()  # Conjunto para armazenar métricas já vistas e evitar duplicatas
        for metric, affinity in metrics_with_affinity:
            if metric not in seen_metrics:
                # Encontrar a categoria correspondente
                for category, metrics in category_to_metrics.items():
                    if metric in metrics:
                        print(f"- {metric} (Afinidade: {affinity:.2f}, Categoria: {category})")
                        seen_metrics.add(metric)  # Adicionar a métrica ao conjunto para evitar duplicatas
                        break
    else:
        print(f"Nenhuma métrica recomendada para o Usuário {user_idx + 1}.")
    print("-" * 40)
```
