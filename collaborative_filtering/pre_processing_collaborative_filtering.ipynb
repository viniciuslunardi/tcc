```python
import pandas as pd
```

```python
df = pd.read_excel('updated_dataset.xlsx')
df.head()
```

```python
df = df.drop(columns=['org_field', 'use_metrics_na', 'agile_methods_split'])
df = df.dropna(subset=['sanitized_metrics'])
df = df.fillna('')
df.head()
```

```python
columns_to_update = ['use_metrics_planning', 'use_metrics_review', 'use_metrics_weekly', 'use_metrics_daily', 'use_metrics_retro']

for column in columns_to_update:
    df[column] = df[column].apply(lambda x: 'Sim' if x == 'Métricas' else 'Não' if x == '' else x)

df.head()
```

```python
metrics_columns = [
    'metric_1', 'metric_2', 'metric_3', 'metric_4', 'metric_5', 'metric_6', 
    'metric_7', 'metric_8', 'metric_9', 'metric_10', 'metric_11', 'metric_12', 
    'metric_13', 'metric_14', 'metric_15', 'metric_16', 'metric_17', 'metric_18'
]

df = df.drop(columns=metrics_columns)
df.head()
```

```python
# Split the 'sanitized_metrics' column by comma and explode it into multiple rows
df = df.assign(sanitized_metrics=df['sanitized_metrics'].fillna('').str.split(', ')).explode('sanitized_metrics')

df.head()
```

```python

# Combine as colunas especificadas em uma única coluna e remova caracteres como ",", "[", "]", "'"
df['combined'] = df[['role', 'years_exp', 'org_size', 'agile_methods', 'use_metrics_planning', 'use_metrics_review', 'use_metrics_weekly', 'use_metrics_daily', 'use_metrics_retro', 'metrics_category']].apply(
    lambda row: ''.join(row.values.astype(str)).replace(' ', '').lower(), axis=1
)

# Remover caracteres como ',', '[', ']', '.' e "'" da coluna 'combined'
df['combined'] = df['combined'].replace(r"[,\[\]\.']", '', regex=True)

df.head()
```

```python
df.to_excel('ready_for_filtering.xlsx', index=False)
```
